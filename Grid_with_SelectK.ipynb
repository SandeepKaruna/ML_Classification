```python
# Import Libraries
import pandas as pd
from sklearn.model_selection import train_test_split
import time
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.feature_selection import RFE
import pickle
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

def SelectKbest(indep_X,dep_Y,n):
    test=SelectKBest(score_func=chi2,k=n)
    fit1=test.fit(indep_X,dep_Y)
    selectK_features=fit1.transform(indep_X)
    return selectK_features

def split_scalar (indep_X,dep_Y):
    x_train,x_test,y_train,y_test=train_test_split(indep_X,dep_Y,test_size=0.25,random_state=0)
    sc=StandardScaler()
    X_train=sc.fit_transform(x_train)
    X_test=sc.transform(x_test)
    return X_train,X_test,y_train,y_test
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB, ComplementNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier

classifiers = [
    {
        'name': 'Support Vector Machine',
        'classifier': SVC(),
        'param_grid': {'C': [0.1, 1, 10], 'gamma': ['scale','auto'], 'kernel': ['linear', 'rbf']}
    },
    {
        'name': 'Random Forest',
        'classifier': RandomForestClassifier(),
        'param_grid': {'n_estimators': [50, 100, 200], 'max_depth': [None, 5, 10]}
    },
    {
        'name': 'Logistic Regression',
        'classifier': LogisticRegression(),
        'param_grid': {'C': [0.1, 1, 10], 'penalty': ['l2'], 'solver': ['liblinear', 'lbfgs']}
    },
    {
        'name': 'Gaussian Naive Bayes',
        'classifier': GaussianNB(),
        'param_grid': {}  # No hyperparameters to tune for Gaussian Naive Bayes
    },
    {
        'name': 'Bernoulli Naive Bayes',
        'classifier': BernoulliNB(),
        'param_grid': {'alpha': [0.1, 1, 10]}
    },
    {
        'name': 'Decision Tree',
        'classifier': DecisionTreeClassifier(),
        'param_grid': {'max_depth': [None, 5, 10], 'min_samples_split': [2, 5, 10]}
    },
    {
        'name': 'K-Nearest Neighbors',
        'classifier': KNeighborsClassifier(),
        'param_grid': {'n_neighbors': [3, 5, 7], 'weights': ['uniform', 'distance']}
    }
]
```

```python
dataset=pd.read_csv("prep.csv")
dataset
```

```python
df2=dataset
df2=pd.get_dummies(df2,drop_first=True)
idp_x=df2.drop('classification_yes',1)
dp_y=df2['classification_yes']
```

```python
K_=5
select_k=SelectKbest(idp_x,dp_y,K_)
X_train,X_test,y_train,y_test=split_scalar(select_k,dp_y)

# Perform grid search for each classifier
Alg=[]
Bst_scr=[]
Accu=[]
for clf in classifiers:
    print(f"Performing Grid Search for {clf['name']}...")
    Alg.append(clf['name'])

    grid_search = GridSearchCV(clf['classifier'], clf['param_grid'], cv=5)
    grid_search.fit(X_train, y_train)

    # Get the best score for each classifier
    best_score = grid_search.best_score_
    Bst_scr.append(best_score)
    
    y_pred=grid_search.best_estimator_.predict(X_test)
    from sklearn.metrics import confusion_matrix
    cm=confusion_matrix(y_test,y_pred)
    
    from sklearn.metrics import accuracy_score
    from sklearn.metrics import classification_report
    
    Accuracy_=accuracy_score(y_test,y_pred)
    report=classification_report(y_test,y_pred)
    Accu.append(Accuracy_)

    # Create DataFrame with best_params and best_score
results_df = pd.DataFrame({
    'Algorithm': Alg,
    'Best Score': Bst_scr,
    'Accuracy': Accu
})
```
