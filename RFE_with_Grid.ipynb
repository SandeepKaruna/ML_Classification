```python
# Import Libraries
import pandas as pd
from sklearn.model_selection import train_test_split
import time
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.feature_selection import RFE
import pickle
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression


def split_scalar (indep_X,dep_Y):
    x_train,x_test,y_train,y_test=train_test_split(indep_X,dep_Y,test_size=0.25,random_state=0)
    sc=StandardScaler()
    X_train=sc.fit_transform(x_train)
    X_test=sc.transform(x_test)
    return X_train,X_test,y_train,y_test
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB, ComplementNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier

classifiers = [
    {
        'name': 'Support Vector Machine',
        'classifier': SVC(),
        'param_grid': {'C': [0.1, 1, 10], 'gamma': ['scale','auto'], 'kernel': ['linear', 'rbf']}
    },
    {
        'name': 'Random Forest',
        'classifier': RandomForestClassifier(),
        'param_grid': {'n_estimators': [50, 100, 200], 'max_depth': [None, 5, 10]}
    },
    {
        'name': 'Logistic Regression',
        'classifier': LogisticRegression(),
        'param_grid': {'C': [0.1, 1, 10], 'penalty': ['l2'], 'solver': ['liblinear', 'lbfgs']}
    },
    {
        'name': 'Gaussian Naive Bayes',
        'classifier': GaussianNB(),
        'param_grid': {}  # No hyperparameters to tune for Gaussian Naive Bayes
    },
    {
        'name': 'Bernoulli Naive Bayes',
        'classifier': BernoulliNB(),
        'param_grid': {'alpha': [0.1, 1, 10]}
    },
    {
        'name': 'Decision Tree',
        'classifier': DecisionTreeClassifier(),
        'param_grid': {'max_depth': [None, 5, 10], 'min_samples_split': [2, 5, 10]}
    },
    {
        'name': 'K-Nearest Neighbors',
        'classifier': KNeighborsClassifier(),
        'param_grid': {'n_neighbors': [3, 5, 7], 'weights': ['uniform', 'distance']}
    }
]
```

```python
def rfe_feature(indep_X,dep_Y,n):
    rfeList=[]
    from sklearn.svm import SVC
    from sklearn.ensemble import RandomForestClassifier
    from sklearn.linear_model import LogisticRegression
    from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB, ComplementNB
    from sklearn.tree import DecisionTreeClassifier
    from sklearn.neighbors import KNeighborsClassifier
    from sklearn.feature_selection import RFE
    
    log_model = LogisticRegression(solver='lbfgs')
    RF = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
       # NB = GaussianNB()
    DT= DecisionTreeClassifier(criterion = 'gini', max_features='sqrt',splitter='best',random_state = 0)
    svc_model = SVC(kernel = 'linear', random_state = 0)
        #knn = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
    rfemodellist=[log_model,svc_model,RF,DT] 
    for i in rfemodellist:
        Log_rfe=RFE(i,n)
        Log_fit=Log_rfe.fit(indep_X,dep_Y)
        Log_feature=Log_fit.transform(indep_X)
        rfeList.append(Log_feature)
    return rfeList
    
        
```

```python
dataset=pd.read_csv("prep.csv")
df2=dataset
df2=pd.get_dummies(df2,drop_first=True)
idp_x=df2.drop('classification_yes',1)
dp_y=df2['classification_yes']
rfe_list=rfe_feature(idp_x,dp_y,5)
```

```python

log_models= ['Logistic Regression','svc_model','RF','DT']
Supp_V_Machine_Ac=[]
Rf_Ac=[]
LR_Ac=[]
GNB_Ac=[]
DT_Ac=[]
KNN_Ac=[]
BNB_Ac=[]


for i in rfe_list:
    X_train,X_test,y_train,y_test=split_scalar(i,dp_y)
    for clf in classifiers:
        print(f"Performing Grid Search for {clf['name']}...")
             
        grid_search = GridSearchCV(clf['classifier'], clf['param_grid'], cv=5)
        grid_search.fit(X_train, y_train)

        # Get the best score for each classifier
        
        y_pred=grid_search.best_estimator_.predict(X_test)
        from sklearn.metrics import confusion_matrix
        cm=confusion_matrix(y_test,y_pred)

        from sklearn.metrics import accuracy_score
        from sklearn.metrics import classification_report

        Accuracy_=accuracy_score(y_test,y_pred)
        report=classification_report(y_test,y_pred)
        if clf['name']=='Support Vector Machine':
            Supp_V_Machine_Ac.append(Accuracy_)
        if clf['name']=='Random Forest':
            Rf_Ac.append(Accuracy_)
        if clf['name']=='Logistic Regression':
            LR_Ac.append(Accuracy_)
        if clf['name']=='Gaussian Naive Bayes':
            GNB_Ac.append(Accuracy_)
        if clf['name']=='Decision Tree':
            DT_Ac.append(Accuracy_)
        if clf['name']=='K-Nearest Neighbors':
            KNN_Ac.append(Accuracy_)
        if clf['name']=='Bernoulli Naive Bayes':
            BNB_Ac.append(Accuracy_)
        
        # Create DataFrame with best_params and best_score
```

```python
data = {
    'Model': log_models,
    'SVM': Supp_V_Machine_Ac,
    'Random Forest': Rf_Ac,
    'Linear Regression': LR_Ac,
    'Gaussian Naive Bayes': GNB_Ac,
    'Decision Tree': DT_Ac,
    'KNN': KNN_Ac,
    'BBernoulli Naive Bayes': BNB_Ac
}

df = pd.DataFrame(data)
```
